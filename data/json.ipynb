```python
# get number of tasks in data that have field 'Review', and separately 'Review_Author'. For each repo, print the number of tasks that have 'Review' and 'Review_Author' fields, and a set of the Review_Authors.

for repo in set(data[i]['repo'] for i in range(len(data))):
    tasks = [data[i] for i in range(len(data)) if data[i]['repo'] == repo]
    review_tasks = [task for task in tasks if 'Review' in task]
    review_author_tasks = [task for task in tasks if 'Review_Author' in task]
    review_authors = set(task['Review_Author'] for task in review_author_tasks)
    print(f'{repo}: {len(tasks)} tasks, {len(review_tasks)} tasks with Review, {len(review_author_tasks)} tasks with Review_Author, {review_authors}')
    # print review for each task, indented
    for task in review_tasks:
        if task['Review']:
            print(f'  {task["instance_id"]}: {task["Review"]}')
```

```python
# look at tasks from repo youtube-dl. remove ones which aren't in this list
good_yt_instance_ids = [23199, 28801 , 29698 , 30582 , 31182 , 31235 , 32725 , 32741 , 32845 , 32987]
good_yt_instance_ids = ['ytdl-org__youtube-dl-' + str(i) for i in good_yt_instance_ids]

yt_tasks = [data[i] for i in range(len(data)) if 'youtube' in data[i]['repo']]

good_yt_tasks = [t for t in yt_tasks if t['instance_id'] in good_yt_instance_ids]

print(len(good_yt_tasks), len(good_yt_instance_ids), [i for i in good_yt_instance_ids if i not in [t['instance_id'] for t in good_yt_tasks]])
```

```python
# remove bad tasks from json
new_data = [t for t in data
            if ('youtube' not in t['repo'])
                or t['instance_id'] in good_yt_instance_ids]
print(len(new_data), len(data), len(yt_tasks) - len(good_yt_tasks))
```

```python
# save new_data to codearena_instances.json
with open('codearena_instances.json', 'w') as f:
    json.dump(new_data, f, indent=2)
```
